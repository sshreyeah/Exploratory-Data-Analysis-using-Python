```python
''' Marks '''
marks=[78,85,90,66,88]
print("maximum marks:",max(marks))
print("minimum marks:",min(marks))
print("total marks:",sum(marks))
print("number of students:",len(marks))
print("average marks:",sum(marks)/len(marks))
print("sorted marks:",marks.sort())
print(marks)
```

```python
''' 2. Grocery Shopping List '''
list=["chocolates","butter","coke","ice cream"]
print("original list:",list)
list.append("apples")
print("list after adding:",list)
list.remove("butter")
print("list after removing:",list)
```

```python
''' 3. A person stores phone contacts in a list '''
contacts= ["rudra","shreya","riddhi","leon"]
print("original list: ",contacts)
contacts.append("vaish")
print ("list after adding contact: ",contacts)
contacts.remove("shreya")
print("list after removing contact: ",contacts)
print("contacts sorted alphabetically: ",contacts.sort())
print(contacts)
```

```python
'''Temperature Analysis'''
temp=[28,20,28,20,25,25,19,17,20]
print("temp for first 5 days: ")
print(temp[0:5])
print("temp for last 5 days: ")
print(temp[-5:])
print("highest temperature: ",max(temp))
print("lowerst temperature: ", min(temp))
print("average temperature: ", sum(temp)/len(temp))
```

```python
''' Daily Expense Tracker'''
expense=[100,200,150,300,250,170,160]
print("total expenses: ",sum(expense))
print("average expenses: ",sum(expense)/len(expense))
print("highest spending day: ",max(expense))
print("lowest spending day: ", min(expense))
```
