# **SETS**

```python
# creation of set
thisset={"math","science","english"}
print(thisset)
```

```python
# set doesn't allow duplicate values
thisset={"math","science","english","math"}
print(thisset)
```

```python
# The values true and 1 are considered equivalent, meaning they are treated as duplicates.
thisset={"math","science","english",True,1}
print(thisset)
```

```python
# A set with strings, integers and boolean values
set={"abc",20,True,40,"male"}
print(set)
```

```python
# to check membership
thisset={"math","science","english"}
print("math" in thisset)
print("hindi" in thisset)
```

```python
# once a set is created, you can not replace its items, but you can add new items.
set1={"mouse","monitor","keyboard"}
set1.add("mousepad")
print(set1)
```

```python
# remove items
set1={"monitor","mouse","keyboard"}
set1.remove("monitor")
print(set1)
```

```python
# set operations
A={1,2,3,4,5}
B={4,5,6,7,8}
print("union: ",A|B)
print("intersection: ",A&B)
print("difference: ",A-B)
print("symmetric difference: ",A^B)
```

```python
# Python Frozenset
x= frozenset({"mango","banana","cherry"})
print(x)
print(type(x))
```

```python
# frozenset does not allow addition or removal of items
x= frozenset({"mango","banana","cherry"})
print(x.add("mango"))
```

# **PYTHON DICTIONARY**

```python
# creating a dictionary
dict={"brand":"ford","model":"mustang","year":"1964"}
print(dict)
```

```python
# dictionary does not allow duplicates, it replaces with the latest value assigned
dict={"brand":"ford","model":"mustang","year":1964,"year":2020}
print(dict)
```

```python
# length of the dictionary
dict={"brand":"ford","model":"mustang","year":"1964"}
print("length of the dictionary: ",len(dict))
print(type(dict))
```

```python
# changing values in a dictionary
car= {"brand":"ford","model":"mustang","year":"1964"}
x=car.values()
car["color"]="red"
print(x)
```
