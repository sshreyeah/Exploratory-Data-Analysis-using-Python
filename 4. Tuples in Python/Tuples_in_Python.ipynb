```python
'''Marks'''
marks_tuple=(78,85,90,66,88)
print("maximum marks:",max(marks_tuple))
print("minimum marks:",min(marks_tuple))
print("total marks:",sum(marks_tuple))
print("number of students:",len(marks_tuple))
print("average marks:",sum(marks_tuple)/len(marks_tuple))
```

```python
list1=(20,30,50,60,40,10,70)
print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-1])
print(list1[6:4:-2])
print(list1[-2:-4:-2])
```

```python
'''Temperature Analysis'''
temp=(28,20,28,20,25,25,19,17,20)
print("temp for first 5 days: ")
print(temp[0:5])
print("temp for last 5 days: ")
print(temp[-5:])
print("highest temperature: ",max(temp))
print("lowerst temperature: ", min(temp))
print("average temperature: ", sum(temp)/len(temp))
```

```python
''' Daily Expense Tracker'''
expense=(100,200,150,300,250,170,160)
print("total expenses: ",sum(expense))
print("average expenses: ",sum(expense)/len(expense))
print("highest spending day: ",max(expense))
print("lowest spending day: ", min(expense))
```
